```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url_vendas = 'https://raw.githubusercontent.com/CauaV1ni/dados-carros-eletricos-br/refs/heads/main/vendas_abve_brasil.csv'
url_emissoes = 'https://raw.githubusercontent.com/CauaV1ni/dados-carros-eletricos-br/refs/heads/main/emissoes_comparativo.csv'

df_vendas = pd.read_csv(url_vendas)
df_emissoes = pd.read_csv(url_emissoes)


plt.figure(figsize=(10, 6))

sns.lineplot(data=df_vendas, x='ano', y='vendas_byd', marker='o', label='Vendas BYD', linewidth=3, color='#00529b')
sns.lineplot(data=df_vendas, x='ano', y='total_eletrificados_brasil', marker='^', label='Total do Mercado (Todas as Marcas)', linewidth=2, color='gray', linestyle='--')

plt.title('Evolução de Vendas: BYD vs Mercado Total de Eletrificados no Brasil', fontsize=14, pad=15)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Unidades Vendidas', fontsize=12)
plt.xticks(df_vendas['ano'])
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend()
plt.show()

print("\n" + "="*60 + "\n")


plt.figure(figsize=(10, 6))

ax = sns.barplot(data=df_emissoes, x='categoria', y='emissao_media_g_km', palette='mako')

plt.title('Emissão Média de CO2 (g/km): Modelos BYD vs Combustão', fontsize=14, pad=15)
plt.xlabel('Categoria / Veículo', fontsize=12)
plt.ylabel('Emissões (gCO2/km)', fontsize=12)


for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points',
                fontweight='bold')

plt.ylim(0, 170)
plt.xticks(rotation=15)

plt.show()
```
